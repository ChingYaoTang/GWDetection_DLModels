```python
def show_result(model1,model2=None,model3=None,model4=None,savedir=None):
    th1 = h5py.File('/Users/tang-ching-yao/Documents/deep_learning/traininghistory/traininghistory-{0}'.format(model1), 'a')
    if model2 != None:
        th2 = h5py.File('/Users/tang-ching-yao/Documents/deep_learning/traininghistory/traininghistory-{0}'.format(model2), 'a')
        accmean = (th1['acc'][:39]+th2['acc'][:39])/2
        senmean = (th1['sen'][:39]+th2['sen'][:39])/2
    if model3 != None:
        th3 = h5py.File('/Users/tang-ching-yao/Documents/deep_learning/traininghistory/traininghistory-{0}'.format(model3), 'a')
        accmean = (th1['acc'][:39]+th2['acc'][:39]+th3['acc'][:39])/3
        senmean = (th1['sen'][:39]+th2['sen'][:39]+th3['sen'][:39])/3
    if model4 != None:
        th4 = h5py.File('/Users/tang-ching-yao/Documents/deep_learning/traininghistory/traininghistory-{0}'.format(model4), 'a')
        accmean = (th1['acc'][:39]+th2['acc'][:39]+th3['acc'][:39]+th4['acc'][:39])/4
        senmean = (th1['sen'][:39]+th2['sen'][:39]+th3['sen'][:39]+th4['sen'][:39])/4
    i=10.

    plt.figure(figsize=(10,13))
    
    ax1 = plt.subplot2grid((5, 1), (0, 0))
    
    plt.plot(th1['pr'][...]*i , th1['acc'][:39],'b',label=model1)
    if model2 != None:
        plt.plot(th2['pr'][...]*i , th2['acc'][:39],'y',label=model2)
    if model3 != None:
        plt.plot(th3['pr'][...]*i,th3['acc'][:39],'k',label=model3)
    if model4 != None:
        plt.plot(th4['pr'][...]*i,th4['acc'][:39],'r',label=model4)
    plt.xticks(np.array([5,6,7,8,9,10,11,12,13,15,20,30,40,50,60,70,80,90,100])/10.*i)
    #plt.xlabel('snr')
    plt.ylabel('accuracy')
    plt.grid('on')
    plt.legend(loc='lower right')
    plt.xlim(0.4*i,1.5*i)
    
    ax2 = plt.subplot2grid((5, 1), (1, 0))

    plt.plot(th1['pr'][...]*i , th1['acc'][:39]-accmean,'b',label=model1)
    if model2 != None:
        plt.plot(th2['pr'][...]*i , th2['acc'][:39]-accmean,'y',label=model2)
    if model3 != None:
        plt.plot(th3['pr'][...]*i, th3['acc'][:39]-accmean,'k',label=model3)
    if model4 != None:
        plt.plot(th4['pr'][...]*i, th4['acc'][:39]-accmean,'r',label=model4)
    plt.xticks(np.array([5,6,7,8,9,10,11,12,13,15,20,30,40,50,60,70,80,90,100])/10.*i)
    #plt.xlabel('snr')
    plt.ylabel('accuracy compare with mean')
    plt.grid('on')
    plt.legend(loc='lower right')
    plt.xlim(0.4*i,1.5*i)
    
    ax3 = plt.subplot2grid((5, 1), (2, 0))
    
    plt.plot(th1['pr'][...]*i , th1['sen'][:39],'b',label=model1)
    if model2 != None:
        plt.plot(th2['pr'][...]*i , th2['sen'][:39],'y',label=model2)
    if model3 != None:
        plt.plot(th3['pr'][...]*i,th3['sen'][:39],'k',label=model3)
    if model4 != None:
        plt.plot(th4['pr'][...]*i,th4['sen'][:39],'r',label=model4)
    plt.xticks(np.array([5,6,7,8,9,10,11,12,13,15,20,30,40,50,60,70,80,90,100])/10.*i)
    #plt.xlabel('snr')
    plt.ylabel('sensitivity')
    plt.grid('on')
    plt.legend(loc='lower right')
    plt.xlim(0.4*i,1.5*i)
    
    ax4 = plt.subplot2grid((5, 1), (3, 0))
    
    plt.plot(th1['pr'][...]*i , th1['sen'][:39]-senmean,'b',label=model1)
    if model2 != None:
        plt.plot(th2['pr'][...]*i , th2['sen'][:39]-senmean,'y',label=model2)
    if model3 != None:
        plt.plot(th3['pr'][...]*i,th3['sen'][:39]-senmean,'k',label=model3)
    if model4 != None:
        plt.plot(th4['pr'][...]*i,th4['sen'][:39]-senmean,'r',label=model4)
    plt.xticks(np.array([5,6,7,8,9,10,11,12,13,15,20,30,40,50,60,70,80,90,100])/10.*i)
    #plt.xlabel('snr')
    plt.ylabel('sensitivity compare with mean')
    plt.grid('on')
    plt.legend(loc='lower right')
    plt.xlim(0.4*i,1.5*i)
    
    ax5 = plt.subplot2grid((5, 1), (4, 0))
    
    plt.plot(th1['pr'][...]*i , th1['far'][:39],'b',label=model1)
    if model2 != None:
        plt.plot(th2['pr'][...]*i , th2['far'][:39],'y',label=model2)
    if model3 != None:
        plt.plot(th3['pr'][...]*i,th3['far'][:39],'k',label=model3)
    if model4 != None:
        plt.plot(th4['pr'][...]*i,th4['far'][:39],'r',label=model4)
    plt.xticks(np.array([5,6,7,8,9,10,11,12,13,15,20,30,40,50,60,70,80,90,100])/10.*i)
    plt.xlabel('snr')
    plt.ylabel('false alarm rate')
    plt.grid('on')
    plt.legend(loc='lower right')
    plt.xlim(0.4*i,1.5*i)
    
    plt.show()
    if savedir != None:
        plt.savefig('/Users/tang-ching-yao/Documents/deep_learning/result/{0}.jpg'.format(savedir))
        
    print 'results of finaltestdata:'
    print ('{0:<22s}: accuracy = {1:8f} , sensitivity = {2:8f} , false alarm rate = {3:8f}'.format(model1, th1['acc'][39], th1['sen'][39], th1['far'][39]))#'plain2 :', 'far =', th1['far'][39], ' sen =', th1['sen'][39], ' acc =',th1['acc'][39]
    th1.close()
    if model2 != None:
        print ('{0:<22s}: accuracy = {1:8f} , sensitivity = {2:8f} , false alarm rate = {3:8f}'.format(model2, th2['acc'][39], th2['sen'][39], th2['far'][39]))
        th2.close()
    if model3 != None:
        print ('{0:<22s}: accuracy = {1:8f} , sensitivity = {2:8f} , false alarm rate = {3:8f}'.format(model3, th3['acc'][39], th3['sen'][39], th3['far'][39]))
        th3.close()
    if model4 != None:
        print ('{0:<22s}: accuracy = {1:8f} , sensitivity = {2:8f} , false alarm rate = {3:8f}'.format(model4, th4['acc'][39], th4['sen'][39], th4['far'][39]))
        th4.close()
```

```python
def show_result_PE(model1,model2=None,model3=None,model4=None,savedir=None):
    th1 = h5py.File('/Users/tang-ching-yao/Documents/deep_learning/traininghistory/traininghistory_PE-{0}'.format(model1), 'a')
    if model2 != None:
        th2 = h5py.File('/Users/tang-ching-yao/Documents/deep_learning/traininghistory/traininghistory_PE-{0}'.format(model2), 'a')
    if model3 != None:
        th3 = h5py.File('/Users/tang-ching-yao/Documents/deep_learning/traininghistory/traininghistory_PE-{0}'.format(model3), 'a')
    if model4 != None:
        th4 = h5py.File('/Users/tang-ching-yao/Documents/deep_learning/traininghistory/traininghistory_PE-{0}'.format(model4), 'a')
    i=10.

    plt.figure(figsize=(10,13))
    
    ax1 = plt.subplot2grid((5, 1), (0, 0))
    
    plt.plot(th1['pr'][...]*i , th1['mape'][:39],'b',label=model1)
    if model2 != None:
        plt.plot(th2['pr'][...]*i , th2['mape'][:39],'y',label=model2)
    if model3 != None:
        plt.plot(th3['pr'][...]*i,th3['mape'][:39],'k',label=model3)
    if model4 != None:
        plt.plot(th4['pr'][...]*i,th4['mape'][:39],'r',label=model4)
    plt.xticks(np.array([5,6,7,8,9,10,11,12,13,15,20,30,40,50,60,70,80,90,100])/10.*i)
    plt.yticks(np.array([0,0.05,0.10,0.15,0.20,0.25]))
    plt.xlabel('snr')
    plt.ylabel('mean absolute percentage error')
    plt.grid('on')
    plt.legend(loc='upper right')
    plt.xlim(0.4*i,5*i)
    plt.show()
    if savedir != None:
        plt.savefig('/Users/tang-ching-yao/Documents/deep_learning/result/{0}.jpg'.format(savedir))
        
    print 'results of finaltestdata:'
    print ('{0:<22s}: mean absolute percentage error = {1:8f}'.format(model1, th1['mape'][39]))#'plain2 :', 'far =', th1['far'][39], ' sen =', th1['sen'][39], ' acc =',th1['acc'][39]
    th1.close()
    if model2 != None:
        print ('{0:<22s}: mean absolute percentage error = {1:8f}'.format(model2, th2['mape'][39]))
        th2.close()
    if model3 != None:
        print ('{0:<22s}: mean absolute percentage error = {1:8f}'.format(model3, th3['mape'][39]))
        th3.close()
    if model4 != None:
        print ('{0:<22s}: mean absolute percentage error = {1:8f}'.format(model4, th4['mape'][39]))
        th4.close()
```
