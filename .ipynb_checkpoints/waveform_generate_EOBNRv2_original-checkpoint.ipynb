```python
# Whitening function.
def whiten(strain, interp_psd, dt):
    Nt = len(strain)
    freqs = np.fft.rfftfreq(Nt, dt)
    #freqs1 = np.linspace(0,4096.,Nt/2+1)

    hf = np.fft.rfft(strain)
    norm = 1./np.sqrt(1./(dt*2))
    white_hf = hf / np.sqrt(np.abs(interp_psd(freqs))) * norm
    white_ht = np.fft.irfft(white_hf, n=Nt)
    
    bb, ab = butter(4, [fband[0]*2./sample_rate, fband[1]*2./sample_rate], btype='band')
    normalization = np.sqrt((fband[1]-fband[0])/(sample_rate/2))
    strain_whitenbp = filtfilt(bb, ab, white_ht) / normalization
    return strain_whitenbp
```

```python
def wgn(multiple,signal,snrdb): #white gaussian noise ,noies長度為signal的multiple倍
    snr = snrdb*1 #snrdb無單位
    #snr = 10**(snrdb/10.) #snrdb單位為分貝(10倍以10為底對數)
    l = len(signal)
    signalpower = np.sum(signal*signal)/l #資料點振幅各自平方再相加，除以資料長度 = 平均功率
    noisepower = signalpower/snr
    return np.random.normal(loc=0.0, scale=1.0, size=multiple*l) * np.sqrt(noisepower) #mean=0,standard deviation=1
```

```python
def wgn2(multiple,signal,snr):
    l = len(signal)
    signalpower = np.max(signal)
    noisepower = signalpower/snr
    return np.random.normal(loc=0.0, scale=1.0, size=multiple*l) * noisepower #mean=0,standard deviation=1
```

```python
def datalenth(minmass,maxmass,fmin):
    mass1 = minmass*1
    mass2 = minmass*1
    #print('%7s%7s%7s%7s%10s'%('mass1','mass2','m1+m2','fl','lenth'))
    #dtype = [('m1',float), ('m2',float), ('m1+m2',float), ('fl',int), ('lenth',int)]
    
    dlist = np.array([[0,0,0,0,0]])#, dtype=dtype)
    
    while mass1 <= maxmass:
        mass2 = mass1
        while mass2 <= maxmass:
            fl = fmin
            l = 10001
            while l > 10000 :
                hp, hc = get_td_waveform(approximant=approximation,
                                     mass1=mass1,
                                     mass2=mass2,
                                     f_lower=fl,
                                     delta_t=1.0/sample_rate)
                del hc
                l = len(hp)
                if  l < 20000:
                    a = np.array([[mass1, mass2, mass1+mass2, fl, l]])#, dtype=dtype)
                    dlist = np.concatenate((dlist,a),axis=0)
                    #print('%8.1f%9.1f%8.1f%7i%10i'%(mass1,mass2,mass1+mass2,fl,l))
                fl = fl+5
            mass2 = mass2 + 1
        mass1 = mass1 + 1
    return dlist
```

```python
def datagenerator(peakratio, datasetsize, filename, gwdata, pdf=None): 
    #計時
    start_time = timeit.default_timer()
    num = datasetsize
    #建立存放資料的檔案
    f = h5py.File(filename,'w')
    #設定存放格式
    f.create_dataset('data', (num,data_length))
    f.create_dataset('label', (num,))
    
    #載入GW data
    g = h5py.File(gwdata,'r')
    waveformamount = len(g['parameter'])
    
    
    if type(peakratio)==float or int :
        for j in range(num):
            #隨機產生0或1的數字，有無訊號的數據比例為1:1
            yesornot = np.random.randint(0,2)
            f['label'][j] = yesornot #將數字加入label的數列中
            
            if yesornot==1: #產生有訊號的數據    
                '''
                m1 = np.random.randint(m1_ini*2, m1_max*2+1)/2 #m1、m2取值5~20，變化間隔為0.5
                m2 = np.random.randint(m2_ini*2, m2_max*2+1)/2 
                total_mass = m1+m2
                if total_mass <= 11 : fl = 45
                elif total_mass <= 14 : fl =40
                elif total_mass <= 17 : fl =35
                elif total_mass <= 23 : fl =30
                elif total_mass <= 34 : fl =25
                elif total_mass <= 40 : fl =20
                elif total_mass <= 55 : fl =15
                elif total_mass <= 105 : fl =10
                hp, hc = get_td_waveform(approximant=approximation,
                                     mass1=m1,
                                     mass2=m2,
                                     f_lower=fl,
                                     delta_t=1.0/sample_rate)
                del hc
                start_index = len(hp) - data_length
                end_index = len(hp)
                hp = hp[start_index:end_index]
                # whitening
                hp = whiten(hp, psd_smooth, dt)
                '''
                nn = np.random.randint(0,waveformamount)
                hp = g['data'][nn]
                
                
                #noise
                hp_max = max(hp)
                noise = np.random.normal(0, 1./peakratio * hp_max, data_length)
                hp_noise = hp + noise
                #將有訊號數據加入signal的數列中 
                f['data'][j] = hp_noise.reshape(1,data_length) 
                
            else: #產生無訊號的數據
                pure_noise = np.random.normal(0, 1, data_length)
                f['data'][j] = pure_noise.reshape(1, data_length) #將無訊號數據加入signal的數列中
            if j%1000==0:
                print(j)
    
    elif type(peakratio)==list :
        if pdf != None:
            sampling = metropolis(num, (peakratio[1]+peakratio[0])/2, pdf)
            i = 1
        for j in range(num):
            #隨機產生0或1的數字，有無訊號的數據比例為1:1
            yesornot = np.random.randint(0,2) 
            f['label'][j] = yesornot
            
            if yesornot==1: #產生有訊號的數據
                #產生隨機peakratio
                if pdf == None:
                    randpeakratio = np.random.sample() * np.abs(peakratio[1]-peakratio[0]) + np.min(peakratio)
                else:
                    randpeakratio = sampling[i]
                    i += 1
                nn = np.random.randint(0,waveformamount)
                hp = g['data'][nn]
                #noise
                noise = np.random.normal(0, 1./randpeakratio * hp_max, data_length)
                hp_noise = hp + noise
                #將有訊號數據加入signal的數列中 
                f['data'][j] = hp_noise.reshape(1,data_length) 
                
            else: #產生無訊號的數據
                pure_noise = np.random.normal(0, 1, data_length)
                f['data'][j] = pure_noise.reshape(1, data_length) #將無訊號數據加入signal的數列中
            if j%1000==0:
                print(j)
    
    f.close()
    stop_time = timeit.default_timer()
    print 'period(s) =', stop_time - start_time
    print('finish')
```

```python
def GWgenerator(mmin, mmax, deltam, filename): 
    #計時
    start_time = timeit.default_timer()
    #建立存放資料的檔案
    f = h5py.File(filename,'w')
    #設定存放格式
    maxmin = (mmax-mmin)/deltam
    num = maxmin*(maxmin+1)/2+(maxmin+1)
    f.create_dataset('data', (num,data_length), maxshape=(None,data_length))
    f.create_dataset('parameter', (num,2), maxshape=(None,2))

    j = 0
    m2 = mmin * 1
    m1 = mmin * 1
    while m1 <= mmax:
        m2 = m1
        while m2 <= mmax:
            total_mass = m1+m2
            if total_mass <= 11 : fl = 45
            elif total_mass <= 14 : fl =40
            elif total_mass <= 17 : fl =35
            elif total_mass <= 23 : fl =30
            elif total_mass <= 34 : fl =25
            elif total_mass <= 40 : fl =20
            elif total_mass <= 55 : fl =15
            elif total_mass <= 105 : fl =10
        
            hp, hc = get_td_waveform(approximant=approximation,
                                     mass1=m1,
                                     mass2=m2,
                                     f_lower=fl,
                                     delta_t=1.0/sample_rate)
            del hc
            start_index = len(hp) - data_length
            end_index = len(hp)
            hp = hp[start_index:end_index]
        
            # whitening
            hp = whiten(hp, psd_smooth, dt)
        
            #Resize
            #if j > 2556 : 
            #    lenth = f['label'].shape[0]
            #    f['data'].resize([lenth+10, data_length])
            #    f['label'].resize([lenth+10, 2])
            f['data'][j] = hp.reshape(1,data_length) 
            f['parameter'][j] = np.array([[m1, m2]])
            
            j += 1
            if j%500==0:
                print(j)
            m2 = m2 + deltam
            
        m1 = m1 + deltam
        
    f.close()
    stop_time = timeit.default_timer()
    print stop_time - start_time,'s'
```

# important sampling

```python
def metropolis(num,initial, pdf):
    xold = initial
    l = []
    while len(l) < num:
        xnew = np.random.random_sample()
        condition = np.random.random_sample()
        if pdf(xnew) >= pdf(xold):
            xold = xnew*1
            l = np.append(l, xold)
        elif pdf(xnew) / pdf(xold) >= condition:
            xold = xnew*1
            l = np.append(l, xold)
    return l
```

# data generator without overlap

```python
def datageneratorwo(peakratio, percentage_of_noise, filename, gwdata, pdf=None): 
    #計時
    start_time = timeit.default_timer()
    #建立存放資料的檔案
    f = h5py.File(filename,'w')
    #載入GW data
    g = h5py.File(gwdata,'r')
    num_waveform = len(g['parameter'])
    num = int(num_waveform/(1-percentage_of_noise))
    #設定存放格式
    f.create_dataset('data', (num,data_length))
    f.create_dataset('label', (num,))
    f.create_dataset('parameter', (num,2))
    #資料筆數的亂數排列
    p = np.random.permutation(num)
    
    
    if type(peakratio)==float or int :
        for j in range(num):
            if p[j] < num_waveform: #有訊號的數據
                f['label'][j] = 1
                f['parameter'][j] = g['parameter'][p[j]]
                hp = g['data'][p[j]]
                #noise
                hp_max = max(hp)
                noise = np.random.normal(0, 1./peakratio * hp_max, data_length)
                hp_noise = hp + noise
                f['data'][j] = hp_noise.reshape(1,data_length) 
                 
            else: #無訊號的數據
                f['label'][j] = 0
                f['parameter'][j] = np.array([0,0])
                pure_noise = np.random.normal(0, 1, data_length)
                f['data'][j] = pure_noise.reshape(1, data_length)
                
    
    elif type(peakratio)==list :
        if pdf != None:
            sampling = metropolis(num_waveform, (peakratio[1]+peakratio[0])/2, pdf)
            i = 0
        for j in range(num):
            if p[j] < num_waveform: #有訊號的數據
                f['label'][j] = 1
                f['parameter'][j] = g['parameter'][p[j]]
                hp = g['data'][p[j]]
                #隨機產生peakratio
                if pdf == None:
                    randpeakratio = np.random.sample() * np.abs(peakratio[1]-peakratio[0]) + np.min(peakratio)
                else:
                    randpeakratio = sampling[i]
                    i += 1
                #noise
                hp_max = max(hp)
                noise = np.random.normal(0, 1./randpeakratio * hp_max, data_length)
                hp_noise = hp + noise
                f['data'][j] = hp_noise.reshape(1,data_length)
                 
            else: #無訊號的數據
                f['label'][j] = 0
                f['parameter'][j] = np.array([0,0])
                pure_noise = np.random.normal(0, 1, data_length)
                f['data'][j] = pure_noise.reshape(1, data_length)
    
    
    f.close()
    g.close()
    stop_time = timeit.default_timer()
    print 'period(s) =', stop_time - start_time
    print('finish')
```
