```python
model1 = 'plain4'
model2 = 'plain4-dilation21'
model3 = 'plain4-pm1'
model4 = 'resnet4'
savedir = 'plain4'

th1 = h5py.File('/Users/tang-ching-yao/Documents/deep_learning/traininghistory/traininghistory-{0}'.format(model1), 'a')
th2 = h5py.File('/Users/tang-ching-yao/Documents/deep_learning/traininghistory/traininghistory-{0}'.format(model2), 'a')
th3 = h5py.File('/Users/tang-ching-yao/Documents/deep_learning/traininghistory/traininghistory-{0}'.format(model3), 'a')
th4 = h5py.File('/Users/tang-ching-yao/Documents/deep_learning/traininghistory/traininghistory-{0}'.format(model4), 'a')
i=1.

plt.figure(figsize=(10,10))
ax1 = plt.subplot2grid((5, 1), (0, 0))

plt.plot(th1['pr'][...]*i , th1['acc'][:39],'b',label=model1)
plt.plot(th2['pr'][...]*i , th2['acc'][:39],'y',label=model2)
plt.plot(th3['pr'][...]*i,th3['acc'][:39],'r',label=model3)
plt.plot(th4['pr'][...]*i,th4['acc'][:39],'k',label=model4)

plt.xticks(np.array([5,6,7,8,9,10,11,12,13,15,20,30,40,50,60,70,80,90,100])/10.)
plt.xlabel('snr')
plt.ylabel('accuracy')
plt.grid('on')
plt.legend(loc='lower right')
plt.xlim(0.4*i,2.*i)


ax2 = plt.subplot2grid((5, 1), (1, 0))

accmean = (th1['acc'][:39]+th2['acc'][:39]+th3['acc'][:39]+th4['acc'][:39])/4
plt.plot(th1['pr'][...]*i , th1['acc'][:39]-accmean,'b',label=model1)
plt.plot(th2['pr'][...]*i , th2['acc'][:39]-accmean,'y',label=model2)
plt.plot(th3['pr'][...]*i, th3['acc'][:39]-accmean,'r',label=model3)
plt.plot(th4['pr'][...]*i, th4['acc'][:39]-accmean,'k',label=model4)

plt.xticks(np.array([5,6,7,8,9,10,11,12,13,15,20,30,40,50,60,70,80,90,100])/10.)
plt.xlabel('snr')
plt.ylabel('accuracy')
plt.grid('on')
plt.legend(loc='lower right')
plt.xlim(0.4*i,2.*i)


ax3 = plt.subplot2grid((5, 1), (2, 0))

plt.plot(th1['pr'][...]*i , th1['sen'][:39],'b',label=model1)
plt.plot(th2['pr'][...]*i , th2['sen'][:39],'y',label=model2)
plt.plot(th3['pr'][...]*i,th3['sen'][:39],'r',label=model3)
plt.plot(th4['pr'][...]*i,th4['sen'][:39],'k',label=model4)

plt.xticks(np.array([5,6,7,8,9,10,11,12,13,15,20,30,40,50,60,70,80,90,100])/10.)
plt.xlabel('snr')
plt.ylabel('sensitivity')
plt.grid('on')
plt.legend(loc='lower right')
plt.xlim(0.4*i,2.*i)


ax4 = plt.subplot2grid((5, 1), (3, 0))

senmean = (th1['sen'][:39]+th2['sen'][:39]+th3['sen'][:39]+th4['sen'][:39])/4
plt.plot(th1['pr'][...]*i , th1['sen'][:39]-senmean,'b',label=model1)
plt.plot(th2['pr'][...]*i , th2['sen'][:39]-senmean,'y',label=model2)
plt.plot(th3['pr'][...]*i,th3['sen'][:39]-senmean,'r',label=model3)
plt.plot(th4['pr'][...]*i,th4['sen'][:39]-senmean,'k',label=model4)

plt.xticks(np.array([5,6,7,8,9,10,11,12,13,15,20,30,40,50,60,70,80,90,100])/10.)
plt.xlabel('snr')
plt.ylabel('sensitivity')
plt.grid('on')
plt.legend(loc='lower right')
plt.xlim(0.4*i,2.*i)


ax5 = plt.subplot2grid((5, 1), (4, 0))

plt.plot(th1['pr'][...]*i , th1['far'][:39],'b',label=model1)
plt.plot(th2['pr'][...]*i , th2['far'][:39],'y',label=model2)
plt.plot(th3['pr'][...]*i,th3['far'][:39],'r',label=model3)
plt.plot(th4['pr'][...]*i,th4['far'][:39],'k',label=model4)

plt.xticks(np.array([5,6,7,8,9,10,11,12,13,15,20,30,40,50,60,70,80,90,100])/10.)
plt.xlabel('snr')
plt.ylabel('false alarm rate')
plt.grid('on')
plt.legend(loc='upper center')
plt.xlim(0.4*i,2.*i)

plt.show()
#plt.savefig('/Users/tang-ching-yao/Documents/deep_learning/result/{0}.jpg'.format(savedir))

print ('%21s : acc = %3.8f , sen = %3.8f , far = %3.8f' % (model1, th1['acc'][39], th1['sen'][39], th1['far'][39]))#'plain2 :', 'far =', th1['far'][39], ' sen =', th1['sen'][39], ' acc =',th1['acc'][39]
print ('%s : acc = %3.8f , sen = %3.8f , far = %3.8f' % (model2, th2['acc'][39], th2['sen'][39], th2['far'][39]))
print ('%18s : acc = %3.8f , sen = %3.8f , far = %3.8f' % (model3, th3['acc'][39], th3['sen'][39], th3['far'][39]))
print ('%20s : acc = %3.8f , sen = %3.8f , far = %3.8f' % (model4, th4['acc'][39], th4['sen'][39], th4['far'][39]))


th1.close()
th2.close()
th3.close()
th4.close()
```
